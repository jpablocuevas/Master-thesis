```python
import matplotlib.pyplot as plt
```

```python
# Setting up the plot

wolf_fig, wolf_plot = plt.subplots(1,1);
wolf_plot.set_xlabel("$x$");
wolf_plot.set_ylabel(r"$\Phi (x)$");
wolf_plot.set_title("Wolf potential plot");
#wolf_plot.set(xlim=(1,6), ylim=(-1, 0.1));
wolf_plot.grid();

# Color array

colors = ["darkgreen", "darkred", "darkblue", "darkorange"];

# Alpha values array 

alpha = ["0.3", "0.5", "0.7", "1"];

for i in range (0,4,1):
    data_ptr = open (str(i)+".txt","r");
    data = data_ptr.readlines();

    # Declaring the x and y arrays for the plot
    W=[];
    x=[];

    for line in data:

        x.append(float(line.split()[0]));
        W.append(float(line.split()[1]));
    
    wolf_plot.plot(x,W,'--',color = colors[i],label = r"$alpha$ = " + alpha[i]);

wolf_plot.legend(loc = "lower right");

    
```
