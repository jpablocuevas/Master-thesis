The integral is

$d = \int_0^σ \left[1-exp(u^{MIE}(r)) \right]$

where

$u^{MIE}(r)=ϵ\left(\left(\frac{\sigma}{r} \right)^{\lambda_r}-\left(\frac{\sigma}{r} \right)^{\lambda_a} \right)$

Using $x=\frac{r}{\sigma}$, $dx=\frac{dr}{\sigma}$ we get

$d*=\frac{d}{\sigma} = \int_0^σ \left[1-exp(-\frac{1}{T*}\left(\frac{1}{x^{\lambda_r}}-\frac{1}{x^{\lambda_a}} \right)) \right] dx$

```python

# import scipy.integrate.quad
from scipy import integrate
T=1.0
f = lambda x: 1-np.exp(-((1/x)**2 -(1/x)**3))
```

```python
def func(L_r,L_a,T_eff):
  return lambda x: 1-np.exp(-(1/T_eff)*((1/x)**L_r -(1/x)**L_a))
```

```python
import scipy.integrate as integrate
import scipy.special as special
```

```python
integrate.quad(func(12,6,0.3), 0, 1)
```

```python
R=np.zeros(len(T_red))
o=0
for i in T_red:
  result=integrate.quad(func(12,6,i), 0, 1)
  R[o]=(result[0])
  o=o+1
```
